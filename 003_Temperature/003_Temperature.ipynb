```python
# Load Variables
from dotenv import load_dotenv

load_dotenv()
```

```python
# Create an API Client
from anthropic import Anthropic
import os

client = Anthropic()
model = "claude-sonnet-4-5"
```

```python
# Funciones auxiliares
from os import system

def add_user_message(messages, user_input):
    messages.append({"role": "user", "content": user_input})

def add_assistant_message(messages, answer):
    messages.append({"role": "assistant", "content": answer})

def chat(messages, system=None, temperature=1.0):
    params = {
        "model": model,
        "max_tokens": 1000,
        "messages": messages,
        "temperature": temperature
    }
    if system:
        params["system"] = system

    response = client.messages.create(**params)
    return response.content[0].text

```

```python
# Lista de mensajes para mantener un historial de la conversación
messages = []
# se obtiene la entrada del usuario 
user_input = input("> ")
print(">", user_input)
# se hace el llamdo a la funcion para agregar el mensaje del usuario a la lista de mensajes
add_user_message(messages, user_input)
#Se hace el llamado al chat de la respuesta del modeloy se guarda en la variable answer utilizando un nivel de temperatura dado
answer = chat(messages,temperature=0.7)
#se hace el llamado a la funcion para agregar el mensaje del asistente a la lista de mensajes
add_assistant_message(messages, answer)
#se imprime en pantalla la respuesta del modelo
print("---")
print(answer)
```
