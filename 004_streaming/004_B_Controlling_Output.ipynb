````python
# Create an API Client
from anthropic import Anthropic
import os

client = Anthropic()
model = "claude-sonnet-4-5"
````

````python
# Funciones auxiliares
from os import system

def add_user_message(messages, user_input):
    messages.append({"role": "user", "content": user_input})

def add_assistant_message(messages, answer):
    messages.append({"role": "assistant", "content": answer})

def chat(messages, system=None, temperature=1.0, stop_sequences=None):
    params = {
        "model": model,
        "max_tokens": 1000,
        "messages": messages,
        "temperature": temperature
    }
    if system:
        params["system"] = system
    if stop_sequences:
        params["stop_sequences"] = stop_sequences

    response = client.messages.create(**params)
    return response.content[0].text
````

````python
import json
# Lista de mensajes para mantener un historial de la conversación
messages = []
#se hace el llamado a la funcion para agregar el mensaje del usuario a la lista de mensajes
add_user_message(messages, "Generate a very short event bridge rule as JSON")
#Se hace el llamado a la función para agregar el mensaje del asistente a la lista de mensajes, 
# con un mensaje que indique que se va a generar un bloque de código JSON que inicia con ```json
add_assistant_message(messages, "```json")
#Se hace el llamado a la función de chat para obtener la respuesta del modelo, 
# utilizando un nivel de temperatura dado y una secuencia de parada que indique el final del bloque de código JSON
#El resultado se guarda en la variabl text
text= chat(messages, temperature=0.7, stop_sequences=["```"])
# Se convierte el texto obtenido en un objeto JSON utilizando la función json.loads, 
# eliminando cualquier espacio en blanco al principio o al final del texto con text.strip()
json.loads(text.strip())
````
