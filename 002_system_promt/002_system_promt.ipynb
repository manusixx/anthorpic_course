```python
# Load Variables
from dotenv import load_dotenv

load_dotenv()
```

```python
# Create an API Client
from anthropic import Anthropic
import os

client = Anthropic()
model = "claude-sonnet-4-5"
```

```python
# Funciones auxiliares
from os import system


def add_user_message(messages, user_input):
    messages.append({"role": "user", "content": user_input})

def add_assistant_message(messages, answer):
    messages.append({"role": "assistant", "content": answer})

def chat(messages, system=None):
    params = {
        "model": model,
        "max_tokens": 1000,
        "messages": messages
    }
    if system:
        params["system"] = system

    response = client.messages.create(**params)
    return response.content[0].text

```

```python
# Lista de mensajes para mantener un historial de la conversación
messages = []
# System Prompt que contiene el rol, tono y tipo de respuesta que queremos obtener de la IA
promt =""" 
    you are a patient math tutor. 
    Do not directly answer a student's question. 
    Guide them to a solution step by step.
    """
user_input = input("> ")
print(">", user_input)

add_user_message(messages, user_input)
answer = chat(messages, system=promt)
add_assistant_message(messages, answer)

print("---")
print(answer)
```
